```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import numpy as np
import datetime
```

# Histograms

### Histogram with number data

```python
figure(figsize=(7, 5), dpi=100)
data = [1, 1, 2, 2, 3, 3, 3, 4, 6, 8, 9, 9]
plt.hist(data, bins=5, range=(0, 8), log=False, color='red')
# plt.hist(data, bins=[0, 4, 5, 7, 8, 10])
plt.show()
```

### Histogram with categorical data

```python
figure(figsize=(7, 5), dpi=100)
data = ['a', 'b', 'c', 'a', 'd', 'f']
plt.hist(data)
plt.show()
```

# Working with dataset

```python
df = pd.read_csv("data/sale_10m.csv", usecols = ["amt_novat", "cod"], dtype = {'cod':'object'})
```

```python
df
```

Number data:

```python
figure(figsize=(7, 5), dpi=100)
plt.hist(df["amt_novat"], bins=5, range=(-20, 20), log=False, color='red')
# plt.hist(df["amt_novat"], bins=[-20, -5, 0, 5, 10, 20])
plt.show()
```

Categorical data:

```python
figure(figsize=(7, 5), dpi=100)
plt.hist(df["cod"])
plt.show()
```

```python
def generate_category_data(value):
    if value > 20:
        return 'High'
    elif value > 5:
        return 'Medium'
    elif value > 0:
        return 'Low'
    else:
        return 'Negative'
    
df['amt_novat_cat'] = df['amt_novat'].apply(generate_category_data)
df
```

```python
figure(figsize=(7, 5), dpi=100)
plt.hist(df["amt_novat_cat"])
plt.show()
```
