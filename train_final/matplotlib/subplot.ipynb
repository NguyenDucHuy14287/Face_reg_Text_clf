```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import numpy as np
import datetime
```

# Multiple subplots in one figure

Add an Axes to the current figure or retrieve an existing Axes.

Subplots can also be generated using subplot() as in the following example:



```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 5.0)
x3 = np.linspace(0.0, 4.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
y3 = np.exp(x3)

figure(figsize=(5, 5), dpi=100) 

plt.subplot(2, 3, (1,2)) # (nrows, ncols, index)
# nrows, ncols: Number of rows and columns of the subplot grid. Both of these are optional with a default value of 1.
# index:  index starts at 1 in the upper left corner and increases to the right. 
#         index can also be a two-tuple specifying the (first, last) indices
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
plt.xlabel('time (s)')

# second subplot
plt.subplot(2, 3, 3)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

# second subplot
plt.subplot(2, 3, 4)
plt.plot(x3, y3, '--')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout(1.0)
plt.show()
```
