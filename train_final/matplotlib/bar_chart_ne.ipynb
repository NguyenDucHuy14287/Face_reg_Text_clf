```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import numpy as np
import datetime
```

# Basic line chart with data

### Vertical bar plot

The bars are positioned at x with the given alignment. Their dimensions are given by height and width. The vertical baseline is bottom (default 0).

```python
bar_name = ['Text', 'Video']
bar_height = [4359, 5252]
bar_height_2 = [4083, 5104]
x = np.arange(len(bar_name))
bar_width = 0.2
y_withd = bar_width

figure(figsize=(7, 5), dpi=100)
plt.bar(x=x - y_withd/2, height=bar_height, width=bar_width, label='None Politic')
plt.bar(x=x + y_withd/2, height=bar_height_2, width=bar_width, label='Politic')

for index, value in enumerate(bar_height):
    plt.text(index - y_withd/2  - 0.05, value + 5, str(value))
for index, value in enumerate(bar_height_2):
    plt.text(index + y_withd/2 - 0.05, value + 5, str(value))
plt.title("Data of Experiment 1")
plt.xticks(ticks = x, labels = bar_name)
plt.legend()
plt.show()
```
